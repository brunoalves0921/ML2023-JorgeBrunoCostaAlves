```python
import pandas as pd
import matplotlib.pyplot as plt

dadosCru = pd.read_excel('Latrocinio.xlsx', sheet_name="2018")
display(dadosCru)

#Fazendo a remoção dos dados duplicados
dados = dadosCru.drop_duplicates()

#Fazendo a remoção dos dados nulos
dados = dados.dropna()

#Colocando os dados em lowercase
# dados['UF'] = dados['UF'].str.lower()
# dados['Município'] = dados['Município'].str.lower()

#plotando um gráfico em pizza para a quantidade de crimes cometidos baseado no sexo do autor
dados['SEXO_PESSOA'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(10,10))
plt.title('Quantidade de crimes cometidos baseado no sexo do autor')
plt.show()

#plotar um gráfico mostrando a quantidade de latrocínios cometidos por mês
dados['MÊS ESTATISTICA'].value_counts().plot.bar(figsize=(10,10))
plt.title('Quantidade de latrocínios cometidos por mês')
plt.show()

#plotar um gráfico mostrando a quantidade de latrocínios cometidos por MUNICIPIO_ELABORACAO
dados['MUNICIPIO_ELABORACAO'].value_counts().plot.bar(figsize=(20,10))
plt.title('Quantidade de latrocínios cometidos por MUNICIPIO_ELABORACAO')
plt.show()









```
